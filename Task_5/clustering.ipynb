```python
sb.set_theme(context="paper", palette = "Blues", style = "whitegrid", font_scale= 1, color_codes= True)
```

```python
pingu.head()
```

```python
x = pingu.iloc[: , [0, 1, 2, 3]].dropna().values
x
```

```python
wcss = []
for i in range(1, 20):
    kmeans = KMeans(n_clusters= i, init = "k-means++", random_state= 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 20), wcss, color = "red")
plt.title("Elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```
