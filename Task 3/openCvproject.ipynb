```python
import cv2
import numpy as np
import math
```

```python
vX_max = 2000.0
vY_max = 2000
v_max = 2000
```

```python
def convertRange(var, min1, max1, min2, max2):
    out_var = (var/(max1-min1))*(max2-min2)
    return out_var
```

```python
def toCoordinate(x, y, theta):
    if (x > 0) and (y < 0):
        f_theta = theta
    elif (x < 0) and (y < 0):
        f_theta = 180 - theta
    elif (x < 0) and (y > 0):
        f_theta = 180 + theta
    else:
        f_theta = 360 - theta
    return f_theta
```

```python
def endPoint(start, x, y):
    x1, y1 = start
    return(x + x1, y1 + y)
```

```python
vid = cv2.VideoCapture(r"c:\Users\Ruchisanjil\Videos\Ball_Tracking.mp4")
x1, y1 = 0, 0
while vid.isOpened():
    ret, frame = vid.read()
    if not ret:
        break
    lower_range = np.array([40, 40, 40])
    upper_range = np.array([80, 255, 255])
    frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    masked_image = cv2.inRange(frame_hsv, lower_range, upper_range)
    contours, hierarchy = cv2.findContours(masked_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if contours:
        x, y, width, height = cv2.boundingRect(max(contours, key=cv2.contourArea))
        centre = (int(x+(width/2)),int(y+(height/2)))
        radius = int((width+height)/4)
        x2, y2 = centre
        print(centre)
        del_x = x2 - x1
        del_y = y2 - y1
        fps = vid.get(cv2.CAP_PROP_FPS)
        t = 1/fps
        p_x = (x2-x1)/t
        p_y = (y2-y1)/t
        p_v = math.sqrt(((p_x)*(p_x) + (p_y)*(p_y)))
        if p_v != 0:
            theta = math.acos(math.fabs(p_x/p_v))
            theta = math.degrees(theta)
        theta = toCoordinate(p_x, p_y, theta)
        v_x = int(convertRange(p_x, 0, vX_max, 0, 250))
        v_y = int(convertRange(p_y, 0, vY_max, 0, 250))
        v_total = int(convertRange(p_v, 0, v_max, 0, 250))    
        endP = endPoint(centre, v_x, v_y) 
        frame = cv2.circle(frame, centre, int((width+height)/4), (255,0,0),2)
        frame = cv2.arrowedLine(frame, centre, (x2 + v_x, y2), (255, 0, 0), 2)
        frame = cv2.arrowedLine(frame, centre, (x2, y2 + v_y), (255, 0, 0), 2)
        frame = cv2.arrowedLine(frame, centre, endP, (0, 0, 255), 2)

        x_text = "X velocity: " + str(p_x) + " p/s"
        y_text = "Y velocity: " + str(p_y) + " p/s"
        v_text = "Velocity: " + f"{p_v: .2f}" + " p/s"
        I_text = "Angle: " + f"{theta: .2f}" + " deg"
        font = cv2.FONT_HERSHEY_COMPLEX
        font_scale = 0.6
        t_color = (0, 0, 0)
        thickness = 1

        cv2.putText(frame, x_text, (610, 20), font, font_scale, t_color, thickness)
        cv2.putText(frame, y_text, (610, 40), font, font_scale, t_color, thickness)
        cv2.putText(frame, v_text, (610, 60), font, font_scale, t_color, thickness)
        cv2.putText(frame, I_text, (610, 80), font, font_scale, t_color, thickness)
        print(f"Circle radius: {radius}\nx direction velocity: {v_x} pixels/sec\ny direction velocity: {v_y} pixels/sec\nTotal speed: {v_total} pixels/sec\nAngle of inclination: {theta} degrees")
        x1 = x2
        y1 = y2
    
    cv2.imshow("Ball Tracking",frame)

    if cv2.waitKey(100) & 0xFF == ord('e'):
        break

vid.release()
cv2.destroyAllWindows()
```
